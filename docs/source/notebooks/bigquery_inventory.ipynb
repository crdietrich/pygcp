## Inventory BigQuery Datasets in GCP Projects  

```python
import pandas as pd
from google.cloud import bigquery

from pygcp import credentials, bq
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.max_colwidth = 1000
```

### 1. Load Google Credentials  

```python
%store -r cred_file
%store -r project_id
%store -r project_list
```

```python
gcp_credentials, project_id = credentials.create_credentials(cred_file=cred_file, 
                                                             project_id=project_id)
```

```python
#project_list = ['bigquery-public-data']
project_list = ['sada-colin-dietrich']
```

```python
verbose = True
data_list = []
for project_id in project_list:
    df_project = bq.inventory(gcp_credentials=gcp_credentials,
                              project_id=project_id,
                              verbose=verbose)
    if df_project is None:
        continue
    df_project['project'] = project_id
    data_list.append(df_project)
    
df = pd.concat(data_list)
df.reset_index(inplace=True)
```

```python
df.info()
```

```python
dfg = df[['size_MB', 'row_count', 'dataset_id', 'project']].groupby(by=['project', 'dataset_id']).sum()
```
