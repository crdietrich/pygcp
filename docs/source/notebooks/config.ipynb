# Configuration for GCP Python API Examples  

To use Google Cloud Platform API libraries, credentials are required to assign expenses and data acesss to the correct billing account. When a virtual machine (VM) is created by a service on GCP environmental variables are set with information needed to access the credentials assigned to the VM. When initialized, a class instance of `google.auth.compute_engine.credentials.Credentials` can be created for authentication.  

Alternatively, access can be granted in a programatic way using service accounts and credential files. This is made within a GCP project and typically exported as a JSON file. To use the credentials a path to the JSON file is passed to the initiating function. Once initialized, a class instance of `google.oauth2.service_account.Credentials` is created for further us in API authentiation.

This notebook sets [IPython Store Magic shared variables](https://ipython.readthedocs.io/en/stable/config/extensions/storemagic.html) to use in other Jupyter Notebooks while keeping passwords and authentication tokens out of those other Notebooks. 


### Alternatives to Jupyter Shared Variables 
1. Move these variables to a Python file, such as `config.py` and import them directly i.e.  
    ```
    >> import config
    >> config.cred_file
    ```
2. Use Google Cloud Secret Manager. When deployed to the cloud, the credential file will not be needed since the VM or service executing the code will already have a service account running it. That service account can be given the needed Identity and Access Management (IAM) role permissions to run the needed API calls. Project ID might be something possible to still include in a `config.py` file as it's not actually providing any access. The Maps API key can be put into Secret Manager which would be need to be a allowed by IAM.
3. Conda environmental variables. Using a conda environment file, environmental variables can be set during environment creation. For example:
    ```
    variables:
      variable_1: variable_data
    ```
4. Set environmental variables using Python. Using the `os` module, set and access variables.
i.e.
    ```
    >> import os
    >> os.environ['VARIABLE_1'] = 'variable data to store'
    >> os.environ['VARIABLE_1']
    'variable data to store'
    ```
