```python
import googlemaps

from dsgcp import maps
```

```python
%store -r maps_api_key
```

```python
gmaps_client = googlemaps.Client(key=maps_api_key)
```

### 1. lookup address from latitude and longitude

```python
lat = 32.869549
lon = -116.973947
```

```python
maps.reverse_geocode(gmaps_client, lat, lon)
```

### 2. lookup latitude and longitude from address

```python
address_1 = '141 Hudson St, Port Townsend, WA 98368'
```

```python
address_2 = '1, 3 Tamanyan St, Yerevan 0009, Armenia'
```

```python
maps.geocode(gmaps_client, address_1, 'us')
```

```python
maps.geocode(gmaps_client, address_2, 'am')
```
