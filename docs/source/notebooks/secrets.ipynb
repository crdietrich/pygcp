```python
import pandas as pd

from dsgcp import credentials
from dsgcp import secret
```

### Load shared variables

```python
%store -r cred_file
%store -r project_id
```

```python
c, p = credentials.create_credentials(cred_file,
                                      project_id=project_id)
```

### List all secret in the Project

```python
secret.list_all(credentials=c, project_id=project_id)
```

### Get Secret Value  

```python
secret.access?
```

```python
secret.access(credentials=c, project_id=project_id, secret_id='test1')
```

### Set, Access and Delete a Secret  

```python
test_secret_id = 'test_A'
```

```python
secret.create?
```

```python
secret.create(credentials=c, 
               project_id=project_id, 
               secret_id=test_secret_id,
               labels={'test_label': 'fun_description'})
```

```python
secret.add_version?
```

```python
secret.add_version(credentials=c,
                    project_id=project_id,
                    secret_id=test_secret_id,
                    payload=f'Test Secret Hello World!'
                   )
```

```python
secret.list_all(credentials=c, project_id=project_id)
```

```python
secret.access(credentials=c,
               project_id=project_id,
               secret_id=test_secret_id
               )
```

```python
secret.delete(credentials=c,
               project_id=project_id,
               secret_id=test_secret_id
               )
```
